```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,precision_score,recall_score
```

```python
from sklearn.tree import DecisionTreeClassifier,plot_tree
```

```python
data=pd.read_csv("/content/iris.csv")
print(data.head())
```

```python
x=data.iloc[:,:-1]
y=data.iloc[:,-1]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.3,random_state=0)
```

```python
clf=DecisionTreeClassifier(criterion="entropy")
clf.fit(x_train,y_train)
```

```python
y_hat=clf.predict(x_test)
print("Accuracy: ",accuracy_score(y_hat,y_test))
print("Precision: ",precision_score(y_hat,y_test,average="macro"))
print("Recall: ",recall_score(y_hat,y_test,average="macro"))
```

```python
tree_plot=plot_tree(clf, feature_names=["sepal.length","sepal.width","petal.length","petal.width"],class_names=["Setosa","Versicolor","Virginica"],filled=True,precision=3)
```
